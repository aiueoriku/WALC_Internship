# 3章最終課題

- KMeans, Dendrogram, DBSCANを比較して、使い方を考察してください

## 各手法の概要
### KMeans
アルゴリズムの流れ
- クラス数を決める
- ランダムにクラスタセンタを割り当てる
- 各データポイントを最も近いクラスタセンタに割り当てる(*)
- クラスタセンタを割り当てられた点の重心に移動
- (*)に戻る
- 収束するまで繰り返し

特徴
- クラスタ数を指定可能⇒クラス数が既知なら有効
- クラスタセンタとの距離が分類基準なので，球形にデータポイントが固まる場合有効

### Dendrogram
- デンドログラムは凝集型クラスタリングの進行過程をツリー状に可視化したもの
- 凝集型クラスタリングはクラスタの統合を繰り返す手法．終了条件はクラスタ数
- クラスタの統合の選択基準は色々ある．例えば，wardは分散が増えないように2つのクラスタを選択する

特徴
- クラスタ数を指定可能
- ツリーの枝の長さからクラスタの分離具合が推測できる

### DBSCAN
- データポイントが密集している箇所を見つける
- 距離eps以内の粒子数がmin_samplesを超えたら，クラスタとして割り当てる
- min_sample以下ならノイズとしてみなす

特徴
- クラスタ数を指定しなくていい
- two-moonsなど複雑な形に強い
- ノイズを検出可能


### 使い方
- まず，pairplot関数などで，データポイントの分布を把握する
- 必要に応じてt-SNEやPCAなどで次元を削減する．スケーリングも使う
- irisのようにクラス数が既知の場合は，kmeansやデンドログラムを採用
- two-moonsのようにクラスタの形が特殊である場合や，クラス数が不明な場合はDBSCANを使用

### 各手法の定量評価

- ARI(adjusted rand index)により，kmeans, ward, DBSCANを比較

```python
iris.calc_all_ARI()
```

wardとkmeansがDBSCANを20から30%程上回る結果となった．

irisはクラス数が既知で，データポイントの分布が特殊な形をしているわけではないため，先の説明通りwardやkmeansが有効であるといえる．
