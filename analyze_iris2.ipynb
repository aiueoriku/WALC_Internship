```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns

class AnalyzeIris:
    def __init__(self): # コンストラクタはインスタンス生成時に自動的に呼び出される
        self.data = None # データを格納する変数を初期化
        
    def get(self): # 引数にselfを入れることで別インスタンスで呼び出せる
        iris = load_iris()
        self.data = pd.DataFrame(iris.data, columns=iris.feature_names)
        self.data["label"] = iris.target
        return self.data

    def get_correlation(self):
        if self.data is None:
            self.get()
        data_without_label = self.data.drop("label", axis=1)
        return data_without_label.corr()
    
    def pair_plot(self, diag_kind="hist"):
        if self.data is None:
            self.get()
        self.data["label"][self.data["label"]==0] = "setosa"
        self.data["label"][self.data["label"]==1] = "sersicolor"
        self.data["label"][self.data["label"]==2] = "virginia"
        return sns.pairplot(self.data, hue="label", diag_kind=diag_kind)
    
```

```python
iris = AnalyzeIris() # インスタンス化
iris.get()
```

```python
iris.get_correlation()
```

```python
iris.pair_plot()
```

```python
iris.pair_plot(diag_kind="kde")
```
